```python
from transformers import *
from summarizer import Summarizer
```

```python
# Load model, model config and tokenizer via Transformers
custom_config = AutoConfig.from_pretrained('google/pegasus-xsum')
custom_config.output_hidden_states=True
custom_tokenizer = AutoTokenizer.from_pretrained('google/pegasus-xsum')
custom_model = AutoModel.from_pretrained('google/pegasus-xsum', config=custom_config)
model = Summarizer(custom_model=custom_model, custom_tokenizer=custom_tokenizer)
```

```python
custom_config = AutoConfig.from_pretrained('bert-base-uncased')
custom_config.output_hidden_states=True
custom_tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
custom_model = AutoModel.from_pretrained('bert-base-uncased', config=custom_config)
model = Summarizer(custom_model=custom_model, custom_tokenizer=custom_tokenizer)
```
